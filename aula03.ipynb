# Conceito fundamental de lista

Uma **lista** em Python é uma **estrutura de dados ordenada, mutável e indexada** que pode armazenar múltiplos valores.

```python
numeros = [10, 20, 30, 40]
```

Características principais:

* **Ordenada** → mantém a ordem de inserção
* **Indexada** → cada elemento possui posição
* **Mutável** → pode ser modificada
* **Heterogênea** → aceita tipos diferentes

```python
dados = [10, "Python", True, 3.14]
```

---

# Lógica de indexação

Cada elemento possui um **índice** que representa sua posição.

```
Lista:  [10, 20, 30, 40]
Indice:  0   1   2   3
```

A lógica de indexação começa em **0**.

```python
lista[0]  # primeiro elemento
lista[2]  # terceiro elemento
```

### Índices negativos

Python permite acessar elementos **de trás para frente**.

```
Lista:  [10, 20, 30, 40]
Indice: -4  -3  -2  -1
```

```python
lista[-1]  # último elemento
lista[-2]  # penúltimo
```

**Lógica mental importante:**

* índices positivos → esquerda → direita
* índices negativos → direita → esquerda

---

# Mutabilidade (conceito lógico importante)

Listas são **mutáveis**, ou seja, podem ser alteradas sem recriar o objeto.

```python
lista = [1, 2, 3]

lista[1] = 10

print(lista)
```

Resultado:

```
[1, 10, 3]
```

Isso significa que a **referência da lista permanece a mesma na memória**.

---

# Estrutura lógica na memória

Uma lista é basicamente:

```
lista → referência → [objeto1, objeto2, objeto3]
```

Cada posição da lista guarda **referências para objetos**, não necessariamente os valores diretamente.

Exemplo:

```python
a = [10, 20]
b = a
```

Agora:

```
a → [10, 20]
b → [10, 20]
```

Se modificar:

```python
b[0] = 99
```

Resultado:

```
a = [99, 20]
b = [99, 20]
```

Porque **a e b apontam para o mesmo objeto**.

# Fatiamento (slicing)

O **slicing** permite acessar subconjuntos da lista.

Sintaxe:

```
lista[inicio:fim:passo]
```

Exemplo:

```python
lista = [0,1,2,3,4,5]

lista[1:4]
```

Resultado:

```
[1,2,3]
```

Regra lógica:

```
inicio → incluído
fim → excluído
```

---

### exemplos importantes

Primeiros elementos:

```python
lista[:3]
```

Últimos elementos:

```python
lista[-3:]
```

Passo:

```python
lista[::2]
```

Resultado:

```
[0,2,4]
```

Lista invertida:

```python
lista[::-1]
```

# Iteração sobre listas

### for tradicional

```python
for item in lista:
    print(item)
```

---

### usando índice

```python
for i in range(len(lista)):
    print(lista[i])
```

---

### usando enumerate

```python
for indice, valor in enumerate(lista):
    print(indice, valor)
```
